```python
%matplotlib inline
from matplotlib import pylab as plt
from skimage import data, io, filters, exposure
from pylab import *
import os
```

```python
working_directory = os.getcwd()
```

```python
def threshold(x, pixels):
    binary = (pixels > x) + 1
    return np.uint8(binary)
```

```python
def filter_image(pixels):
    return filters.sobel(threshold(0.5, pixels))
```

```python
for i in range(0, 21):
    img_number = f'{i:02}'
    pixels = data.load(working_directory + '/img/samolot' + img_number + '.jpg', True)
    
    # set size for an image
    figure(figsize=(20,200))
    
    # select odd subplots
    subplot(21,2, ((i + 1) * 2) - 1)
        
    # draw to the selected subplots
    io.imshow(pixels)
    
    # select even subplots
    subplot(21,2, (i + 1) * 2)
    
    # draw to the selected subplots
    io.imshow(filter_image(pixels))
```
